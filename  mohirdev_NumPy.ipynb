#### 19. Tasodifiy sonlardan iborat bo'lgan massiv yarating, so'ng ularni quyida berilgan 4 xil usul yordamida musbat sonlarning butun qismini ajratib oling.

`ðŸ’¡Ishora` %, np.floor, astype, np.trunc

````python
# Create a random array of floats
random_array = np.random.rand(5)

# Method 1: Using the modulo operator
positive_integers = random_array % 1

# Method 2: Using np.floor
positive_integers = np.floor(random_array)

# Method 3: Using astype(int)
positive_integers = random_array.astype(int)

# Method 4: Using np.trunc
positive_integers = np.trunc(random_array)

# Print the results
print("Original array:", random_array)
print("Positive integers (method 1):", positive_integers)
print("Positive integers (method 2):", positive_integers)
print("Positive integers (method 3):", positive_integers)
print("Positive integers (method 4):", positive_integers)
````

#### 25. NumPy kutubxonasi yordamida kecha, bugun va ertangi kun sanalarini aniqlang.

````python
# Get today's date
today = np.datetime64('today')

# Calculate yesterday's date
yesterday = today - np.timedelta64(1, 'D')

# Calculate tomorrow's date
tomorrow = today + np.timedelta64(1, 'D')

# Print the dates
print(f"Yesterday: {yesterday}")
print(f"Today: {today}")
print(f"Tomorrow: {tomorrow}")
````

#### 28. Ixtiyoriy kichik massivlar yarating, ular yeg'indisini np.sum funksiyasidanda tezroq hisoblovchi dastur tuzing.


`ðŸ’¡Ishora` np.add.reduce

````python
# Create two small arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Calculate the sum using np.sum
sum_using_np_sum = np.sum(a) + np.sum(b)

# Calculate the sum using np.add.reduce
sum_using_np_add_reduce = np.add.reduce([a, b])

# Print the results
print(f"Sum using np.sum: {sum_using_np_sum}")
print(f"Sum using np.add.reduce: {sum_using_np_add_reduce}")
````

#### 31. Ixtiyoriy sonlardan tashkil topgan massiv yarating, unga o'xshash qiymatlar kiriting. So'ng, o'xshash qiymatlarni topuvchi dastur tuzing.


`ðŸ’¡Ishora` np.bincount, argmax

````python
# Create an array with some duplicate values
array = np.array([1, 2, 2, 3, 4, 4, 5, 5, 5])


counts = np.bincount(array)

# Find the index of the most frequent value
most_frequent_index = np.argmax(counts)

# Get the most frequent value
most_frequent_value = array[most_frequent_index]

# Print
print(f"Most frequent value: {most_frequent_value}")
print(f"Number of times the most frequent value occurs: {counts[most_frequent_index]}")
````
